```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("amazon.csv")
df
```

```python
df.columns
```

```python
#Q.1.What is the average rating for each product category?
```

```python
df["rating"] = df["rating"].replace('|', float('nan')).astype("float64")
df["rating"]
```

```python
#Average rating for each category:
average_rating=df.groupby("category")["rating"].mean()
```

```python
average_rating.round(2)
```

```python
#If overall average is required:
print("Overall average rating:",average_rating.mean())
```

```python
#Que.2.What are the top rating_count products by category?
df["rating_count"].fillna(0, inplace=True)
```

```python
df["rating_count"].replace(",", "", regex=True, inplace=True)
# df["rating"] = df["rating"]
```

```python
df["rating_count"].astype("int64")
```

```python
df["rating_count"] = df["rating_count"].astype(int)
df_2 = df[df["rating_count"].notnull()]
```

```python
df_2
```

```python
#top 10 rating_count products by rating_counts:
df_2[["category","product_name","rating_count"]].nlargest(10,"rating_count")
```

```python
df_2[["category","product_name","rating_count"]].sort_values(by="rating_count",ascending=False)
```

```python
#top 10 rating_count products by category:
top_10_products = df_2.groupby("category").apply(lambda x: x.nlargest(10, "rating_count")).reset_index(drop=True)
top_10_products = top_10_products[["category", "product_name", "rating_count"]]
top_10_products.sort_values(by="rating_count",ascending=False)[0:20]
```

```python
#Que.3.What is the distribution of discounted prices vs. actual prices?
```

```python
df[["discounted_price","actual_price"]]
```

```python
DP=df["discounted_price"].str.replace("₹","").str.replace(",","").astype("float")
DP
```

```python
AP = df["actual_price"].str.replace("₹","").str.replace(",","").astype("float")
AP
```

```python
#4.How does the average discount percentage vary across categories?
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace("%", "").astype(float)
average_discount_by_category = df.groupby("category")["discount_percentage"].mean()
```

```python
average_discount_by_category.sort_values(ascending=False)
```

```python
more_than_average = (average_discount_by_category >= average_discount_by_category.mean()).sum()
less_than_average= (average_discount_by_category < average_discount_by_category.mean()).sum()
print("more_than_average:",more_than_average)
print("less_than_average:",less_than_average)
```

```python
"""Insight:Hence as per the study, there are 115 out of 211 categories which are having "average discount percentage"
greater than or equal to overall mean viz: 47.69, while on the other hand, there are 96 categories out of 211, which
are having less "average discount percentage" than 47.69 (overall mean)."""
```

```python
#6. What are the most popular product keywords?
```

```python
# Assuming your DataFrame is named 'amazon_sales_data'
keywords = df["product_name"].str.cat(df["about_product"],sep=" ",na_rep="").str.lower().str.split()
#Flatten the list
word_list=[word for sublist in keywords for word in sublist]
#Count the frequency of words
word_count=pd.Series(word_list)
word_count.value_counts()
#Top 20 most popular product keywords used are:
top_keywords=word_count.value_counts().head(20)
print(top_keywords)
```

```python
#8.What is the correlation between discounted_price and rating?
```

```python
df.head(1)
```

```python
import warnings
df_8=df[["discounted_price","rating"]]
df_8["discounted_price"]=df_8["discounted_price"].str.replace("₹","").str.replace(",","").astype(float)
df_8
```

```python
df_8.corr()
```

```python
sns.scatterplot(x=df_8["discounted_price"],y=df_8["rating"],color="r")
plt.title("Correlation between Rating and Discounted_price")
plt.show()
```

```python
'''Insight and Conclusion: As per the above studies suggest that Rating and Discounted_price has very weak positive
correlation of 0.12, which suggest that as the rating increases the increase in discounted_price also happens slightly,
But a the correlation is not so strong we should keep other factors in mind too.
There may be other factors such as product quality, brand reputation, or market demand that influence both the rating
and the discounted price of a product. So, it's always a good idea to consider multiple factors before drawing any
conclusions.'''
```
