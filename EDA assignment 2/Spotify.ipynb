```python
'''
1.Identify the top 5 popular artists based on the mean popularity of their tracks.Show the mean popularity of
tracks for the top 5 popular artists Using BarPlot.
'''
```

```python
df["Artist"].value_counts()
```

```python
df_1=df[["Artist","Popularity"]].groupby("Artist").mean().round(2)
Top_Popular_Artist=df_1.reset_index().sort_values(by="Popularity",ascending=False)
Top_Popular_Artist
```

```python
top_5_Popular_Artist=Top_Popular_Artist.head(5)
top_5_Popular_Artist
```

```python
#Plot a bar
plt.figure(figsize=(8,4))
sns.barplot(x=top_5_Popular_Artist["Artist"],y=top_5_Popular_Artist["Popularity"],width=0.5,color="red")
for i, count in enumerate(top_5_Popular_Artist["Popularity"]):
  plt.text(i,count,str(count),ha="center",va="bottom")
plt.title("mean popularity of tracks for the top 5 popular artists Using BarPlot")
plt.show()
```

```python
'''
2. Determine the top 5 popular songs based on their popularity ratings.Display the popularity ratings of the top
5 popular songs using BarPlot.
'''
```

```python
df_2=df[["Track Name","Popularity"]]
df_2
```

```python
Top_popular_songs=df_2.sort_values(by="Popularity",ascending=False).drop_duplicates()
Top_popular_songs
```

```python
Top_5_popular_songs=Top_popular_songs.head(5)
Top_5_popular_songs
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=Top_5_popular_songs["Track Name"],y=Top_5_popular_songs["Popularity"],color="r",width=0.5)
plt.xticks(rotation=5)
for i,count in enumerate(Top_5_popular_songs["Popularity"]):
  plt.text(i,count,str(count),va="bottom",ha="center")
plt.show()
```

```python
'''
3.Find the top 5 trending genres based on the mean popularity of tracks within each genre.Visualize the mean
popularity of tracks for the top 5 trending genres.
'''
#Since, there is no specified genre defined,let's follow the different approach
mean_popularity_by_artist = df.groupby('Artist')['Popularity'].mean().reset_index()
mean_popularity_by_artist
     
```

```python
top_5_artists = mean_popularity_by_artist.nlargest(5, 'Popularity').round(2)
top_5_artists
```

```python
#Visualize the mean popularity of tracks for the top 5 artists
plt.figure(figsize=(6, 4))
sns.barplot(x='Artist', y='Popularity', data=top_5_artists, color='r',width=0.5)
plt.xlabel('Artist')
plt.ylabel('Mean Popularity')
plt.title('Mean Popularity of Tracks for Top 5 Artists')
for i,count in enumerate(top_5_artists["Popularity"]):
  plt.text(i,count,str(count),ha="center",va="bottom")
plt.show()
```

```python
'''
4.Identify the top 5 longest songs among the tracks of the top 5 popular artists.Represent the duration of the
top 5 longest songs among the tracks of the top 5 popular artists using BarPlot.
'''
```

```python
mean_popularity_by_artist = df.groupby('Artist')['Popularity'].mean().reset_index()
mean_popularity_by_artist
     
```

```python
top_5_artists = mean_popularity_by_artist.nlargest(5, 'Popularity')
top_5_artist_tracks =df[df["Artist"].isin(top_5_artists['Artist'])].drop_duplicates()
top_5_artist_tracks
```

```python
#Sorting the tracks by duration and selecting the top 5 longest songs
top_5_longest_songs = top_5_artist_tracks.nlargest(5, 'Duration (ms)')
top_5_longest_songs
```

```python
#Visualizing the duration of the top 5 longest songs using a bar plot
plt.figure(figsize=(8, 4))
sns.barplot(x='Track Name', y='Duration (ms)', data=top_5_longest_songs, color='r',width=0.5)
plt.title('Duration of Top 5 Longest Songs')
plt.xticks(rotation=15)
for i,count in enumerate(top_5_longest_songs["Duration (ms)"]):
  plt.text(i,count,str(count),ha="center",va="bottom")
plt.show()
```

```python
'''
5.Determine the top 5 most danceable songs based on their danceability scores.Illustrate the danceability
scores of the top 5 most danceable songs using PieChart.
'''
```

```python
df_numerical = df[['Popularity', 'Duration (ms)']]

# Create the pair plot
sns.pairplot(df_numerical)
plt.show()
```

```python
# Create the violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='Artist', y='Duration (ms)', data=df)
plt.xticks(rotation=45)
plt.title('Track Duration Distribution Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.show()
```

```python
"""The plot shows that while some artists consistently produce tracks within a similar range of durations, others experiment with a wider range of track lengths. This variation could reflect differences in artistic style, album structure, or genre preferences."""
```

```python
# Create the swarm plot
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=45)
plt.title('Track Popularity Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.show()
```

```python
'''High Popularity Artists show consistently high popularity, with fewer low-performing tracks.
Variable Popularity Artists exhibit greater variability in their track popularity, indicating both hits and less popular releases.
Mid-Level Popularity Artists tend to have a majority of tracks that fall in the mid-range of popularity, indicating consistent but not top-level performance (e.g., DJ Khaled, Jaden Smith).'''
```
